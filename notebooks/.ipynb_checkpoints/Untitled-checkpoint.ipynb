```python
class MyStrategy(MFW_Simulation):
    
    def tradingStrategy(self,St,Wt,Bt,delta_t):
        if len(St) < 1:
            return 0
        if St[-1] > St[0]:
            return 1
        elif St[-1] == St[0]:
            return 0
        else:
            return -1
```
