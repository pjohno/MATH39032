```python
import matplotlib.pyplot as plt
import scipy.stats as si
import numpy as np
```

```python
T = 1.
timeSteps = 100
```

```python
def stockPath(S0,mu,sigma,T,timeSteps):
    St = np.zeros(timeSteps+1)
    dt = float(T) / float(timeSteps)
    St[0] = S0
    for i in range(timeSteps):
        St[i+1] = St[i] * np.exp( (mu - 0.5*sigma*sigma) *dt + np.sqrt(dt)*sigma*np.random.normal() )
    return St
```

```python
# create a figure with two subplots
fig = plt.figure(figsize=(16, 12), dpi=80)
ax1 = plt.subplot(2,2,1 )
ax2 = plt.subplot(2,2,2 )
ax3 = plt.subplot(2,2,3 )
ax4 = plt.subplot(2,2,4 )

# store stats
simulationRuns = 10
# get the time axes for the bottom
t = np.linspace(0, T, num=(timeSteps+1))

for i in range(simulationRuns):
    St_1 = stockPath(1.,0.2,0.5,T,timeSteps)
    St_2 = stockPath(2.,0.1,0.1,T,timeSteps)
    St_3 = stockPath(1.,-0.1,0.2,T,timeSteps)
    St_4 = stockPath(0.5,-0.2,0.01,T,timeSteps)
    # generate a plot of the stock price 
    ax1.plot(t, St_1 )
    # generate a plot of the stock price 
    ax2.plot(t, St_2 )
    # generate a plot of the stock price 
    ax3.plot(t, St_3 )
    # generate a plot of the stock price 
    ax4.plot(t, St_4 )
```
